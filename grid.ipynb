```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Grid size
rows, cols = 5, 5

# Create a figure
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the grid
for i in range(rows + 1):
    ax.plot([0, cols], [i, i], color='black', linewidth=0.8)
for j in range(cols + 1):
    ax.plot([j, j], [0, rows], color='black', linewidth=0.8)

# Add staggered components
for i in range(rows):
    for j in range(cols):
        # Pressure points at cell centers
        ax.text(j + 0.5, rows - i - 0.5, 'P', color='blue', ha='center', va='center', fontsize=12)

# Velocity components
for i in range(0, rows):
    for j in range(0, cols):
        # Horizontal velocity (u) at edges
        if j <= cols:
            ax.scatter(j, rows - i - 0.5, color='green', s=300, zorder=3)  # Disk for u
            ax.text(j, rows - i - 0.5, 'u', color='white', ha='center', va='center', fontsize=10, zorder=4)
        # Vertical velocity (v) at edges
        if i <= rows:
            ax.scatter(j + 0.5, rows - i, color='red', s=300, zorder=3)  # Disk for v
            ax.text(j + 0.5, rows - i, 'v', color='white', ha='center', va='center', fontsize=10, zorder=4)

# One extra
for j in range(0, rows):
    ax.scatter(cols, rows - j - 0.5, color='green', s=300, zorder=3)  # Disk for v
    ax.text(cols, rows - j - 0.5, 'u', color='white', ha='center', va='center', fontsize=10, zorder=4)
    
for i in range(0, cols):
    ax.scatter(i + 0.5, 0, color='red', s=300, zorder=3)  # Disk for v
    ax.text(i + 0.5, 0, 'v', color='white', ha='center', va='center', fontsize=10, zorder=4)


# Label the axes
ax.set_xticks(np.arange(0, cols + 1, 1))
ax.set_yticks(np.arange(0, rows + 1, 1))
ax.set_xlim(-0.5, cols + 0.5)
ax.set_ylim(-0.5, rows + 0.5)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('X-direction')
ax.set_ylabel('Y-direction')

# Remove axes for clarity
ax.axis('off')
plt.title("Staggered Grid for Velocity and Pressure")

# Save or show the diagram
# plt.show()
plt.savefig('assets/staggered_grid.png', dpi=300, bbox_inches='tight')
```

```python
# Grid size
rows, cols = 5, 5

# Create a figure
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the grid
for i in range(rows + 1):
    ax.plot([0, cols], [i, i], color='black', linewidth=0.8)
for j in range(cols + 1):
    ax.plot([j, j], [0, rows], color='black', linewidth=0.8)

# Add non-staggered components
for i in range(rows + 1):  # Loop over all grid points
    for j in range(cols + 1):
        # Add a filled circle to represent colocated variables
        ax.scatter(j, rows - i, color='blue', s=700, zorder=3)  # Disk for colocated variables
        # Add the text for each variable
        ax.text(j, rows - i, 'p/u', color='white', ha='center', va='center', fontsize=10, zorder=4)

# Label the axes
ax.set_xticks(np.arange(0, cols + 1, 1))
ax.set_yticks(np.arange(0, rows + 1, 1))
ax.set_xlim(-0.5, cols + 0.5)
ax.set_ylim(-0.5, rows + 0.5)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('X-direction')
ax.set_ylabel('Y-direction')

# Remove axes for clarity
ax.axis('off')
plt.title("Non-Staggered Grid for Velocity and Pressure")


# Save or show the diagram
# plt.show()
plt.savefig('assets/non_staggered_grid.png', dpi=300, bbox_inches='tight')
```
